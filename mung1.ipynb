```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

file="output/education_data.csv"
cx_df=pd.read_csv(file)

temp_df=cx_df.loc[cx_df["year"]>=2003]

data_df=temp_df.loc[temp_df["year"]<=2012]
data_df.head()
```

```python

year_range=['2003','2004','2005','2006','2007','2008','2009','2010','2011','2012']    
region_df=pd.DataFrame(year_range,columns=["year"])
education_df=pd.DataFrame(year_range,columns=["year"])
origin_df=pd.DataFrame(year_range,columns=["year"])
age_df=pd.DataFrame(year_range,columns=["year"])
area_df=pd.DataFrame(year_range,columns=["year"])
```

```python
#region group -- group 8,9,10,11
for x in range(8,12,1):
    select_df=data_df.loc[data_df["group"]==x,:] 
    group_name= select_df.iloc[0,0]
    region_df[group_name]=select_df["value"].values
region_df
```

```python
outfile='output/region.csv'
region_df.to_csv(outfile,index=False,header=True)
```

```python
#education group -- group 12,13,14,15,16,17,18,19
for x in range(12,20,1):
    select_df=data_df.loc[data_df["group"]==x,:] 
    group_name= select_df.iloc[0,0]
    education_df[group_name]=select_df["value"].values
education_df
```

```python
outfile='output/education.csv'
education_df.to_csv(outfile,index=False,header=True)
```
